```python
columns = ["X0", "X1", "X2", "X3", "X4", "X5", "X6"]
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/mercedes.csv',
                 usecols=columns)
```

```python
categories_count = []
length_of_columns = len(df1.columns)
for i in df1.columns:
    if len(df1[i].unique()) >= 10:
        categories_count.append(10)
    else:
        categories_count.append(len(df1[i].unique()))
        
print(categories_count)
print(sum(categories_count) + length_of_columns)
```

```python
for i in df1.columns:
    list1 = list(df1[i].value_counts().sort_values(ascending = False)[:10].index)
    print(list1)
```

```python
for i in df1.columns:
    list1 = list(df1[i].value_counts().sort_values(ascending = False)[:10].index)
    for j in list1:
        df1[i+"_"+j] = np.where(df1[i] == j, 1, 0)
```
